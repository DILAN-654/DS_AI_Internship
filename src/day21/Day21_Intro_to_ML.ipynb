# TASK 1 - The Paradigm Shift (Traditional vs. Machine Learning)

Machine Learning is a way of building computer systems that can learn patterns from data instead of being programmed with fixed instructions. In traditional programming, we provide the computer with rules and input data to get results, but in machine learning we give the system data along with expected outputs, and it automatically learns the rules needed to make predictions on new data. There are different types of machine learning based on how the system learns. In Supervised Learning, the model learns from labeled data, such as predicting house prices based on past price data or detecting spam emails using examples marked as spam or not spam. In Unsupervised Learning, the model finds patterns in data without labels, like grouping customers into different segments based on their shopping behavior. In Reinforcement Learning, the system learns by trial and error using rewards and penalties, such as a robot learning to walk or an AI learning to play games by improving its performance over time.

# TASK 2 Types of Machine Learning

Machine Learning is broadly divided into three main branches:

-   Supervised Learning
-   Unsupervised Learning
-   Reinforcement Learning

Each branch differs based on how data is provided and how learning
happens.

## 1. Supervised Learning

Supervised Learning uses labeled datasets, where each input has a known
correct output.  
The model learns relationships between features and targets to predict
outcomes for new data.  
It is mainly applied to classification and regression problems.

### Examples

-   Spam Email Detection  
-   House Price Prediction  
-   Student Performance Prediction  
-   Disease Diagnosis

## 2. Unsupervised Learning

Unsupervised Learning works with unlabeled data, meaning no predefined
answers exist.  
The algorithm discovers hidden structures, similarities, or clusters
automatically.  
It is commonly used for pattern discovery and data exploration.

### Examples

-   Customer Segmentation  
-   Market Basket Analysis  
-   Fraud Detection  
-   News Article Grouping

## 3. Reinforcement Learning

Reinforcement Learning trains an agent through interaction with an
environment using rewards and penalties.  
The system learns optimal actions via trial and error to maximize
cumulative rewards.  
It is suitable for decision-making and sequential learning problems.

### Examples

-   Self-Driving Cars  
-   Game Playing AI (Chess/Go)  
-   Robot Navigation  
-   Recommendation Optimization Systems

## Structured Comparison Table

| Feature | Supervised Learning | Unsupervised Learning | Reinforcement Learning |
|---------|---------------------|----------------------|------------------------|
| Data Type | Labeled Data | Unlabeled Data | Reward-Based Feedback |
| Learning Method | Learn from known answers | Discover hidden patterns | Learn via rewards & penalties |
| Primary Goal | Prediction | Pattern Discovery | Decision Optimization |
| Human Guidance | High | Low | Medium (environment feedback) |
| Common Tasks | Classification, Regression | Clustering, Association | Control & Decision Making |
| Example 1 | Spam Detection | Customer Segmentation | Self-Driving Cars |
| Example 2 | House Price Prediction | Market Basket Analysis | Game Playing AI |
| Typical Algorithms | Linear Regression, Random Forest | K-Means, PCA | Q-Learning, DQN |

## Why This Matters

Choosing the correct learning paradigm is the first step in Machine
Learning.  
A wrong selection leads to poor model performance even before coding
begins.

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Load dataset
df = pd.read_csv("D:/DS_AI_Internship/data/canada_per_capita_income.csv")
print(df.head())
```

```python

# Training data
X = df[['year']]
y = df['per capita income (US$)']

# Train model
model = LinearRegression()
model.fit(X, y)
```

```python
# Predict for 2020
print("Income in 2020:", model.predict([[2020]]))

# Predict 2020–2030
years = np.arange(2020, 2031).reshape(-1,1)
predictions = model.predict(years)

for year, income in zip(years, predictions):
    print(year[0], ":", income)
```

```python
# Plot graph
plt.scatter(X, y)
plt.plot(X, model.predict(X), color='red')
plt.xlabel("Year")
plt.ylabel("Income")
plt.show()
```
